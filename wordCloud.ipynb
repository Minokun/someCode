```python
import jieba
import re
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# from sqlalchemy import create_engine
# import pandas as pd

def wordClound(str_article):
    del_list_compile = re.compile(r"[，|。|、|\n|”|“|的|和]")
    str_list = jieba.cut(re.sub(del_list_compile, '', str_article))
    # pd.Series(str_list).value_counts().sort_values(ascending=False)
    text = ' '.join(str_list)

    wordcloud = WordCloud(font_path='/System/Library/Fonts/STHeiti Light.ttc', max_font_size=40, background_color='white').generate(str_article)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")

    # lower max_font_size
    wordcloud = WordCloud(max_font_size=80).generate(text)
    plt.figure(figsize=(300, 300))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.savefig("/Users/monstar/Desktop/examples.jpg")
    plt.show()
    
def getDBConnect():
    mysql_setting = {
        'host' : "localhost",
        'dbname': "spider",
        'user': "root",
        'password': ""
    }
    engine = create_engine('mysql://{user}:{password}@{host}:3306/{dbname}'.format(**mysql_setting), echo=False)
    conn = engine.connect()
    return conn

# sql = '''
#     select advantage from 51job_position where advantage <> 'NULL'
# '''
# conn = getDBConnect()
# res_df = pd.read_sql(sql, conn)
```
