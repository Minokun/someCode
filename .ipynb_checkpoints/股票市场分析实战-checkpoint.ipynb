```python
ts_fc = ts.forecast_data(2017,4)
```

```python
ts_fc= ts_fc.set_index('code')
```

```python
df_ts_gn = ts.get_concept_classified()
code_filter = df_ts_gn[df_ts_gn.c_name.isin(df_ts_gn[df_ts_gn.code=='600460'].c_name.values)].code.values
```

```python
# ts_today[ts_today.code.isin(code_filter)].sort_values('changepercent', ascending=False)
def judge_range(line):
    return pd.Series([round(line, 0)])
ts_today['judge'] = ts_today.changepercent.apply(judge_range)
```

```python
sns.set()
ts_today.groupby('judge').size().sort_index().plot(kind='bar', figsize=(16,8), fontsize=14)
```

```python
labels = ts_today.groupby('judge').size().index
sizes = ts_today.groupby('judge').size().values
explode = tuple(np.zeros(len(sizes)))  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots(figsize=(16, 16))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
