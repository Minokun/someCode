```python
import time
from datetime import datetime 
today = datetime.now().strftime("%Y%m%d")
today
```

```python
def getTR(row):
    return max(abs(float(row['high']) - float(row['low'])), abs(float(row['high']) - float(row['pre_close'])), abs(float(row['low']) - float(row['pre_close'])))

def getTRRatio(row):
    return round(abs(row['TR'] /  float(row['pre_close'])) * 100, 2)
```

```python
import seaborn as sns
sns.set()
```

```python
def showRes(df, title, days=14):

    df['TR'] = df.apply(lambda row: getTR(row), axis = 1)
    df['TRP'] = df.apply(lambda row: getTRRatio(row), axis = 1)
    df['ATR'] = round(df['TR'].rolling(window=days).mean(), 2)
    df_deal = df[df['ATR'] > 0]
    df_deal.head()
    
    plt.figure(figsize=(16, 12))
    
    plt.title(title)
    
    #添加标签
    for a, b in list(zip(df_deal['trade_date'], df_deal['pct_chg'])):
        plt.text(a, b + 0.5 * (-1 if b <= 0 else 1), str(b), ha='center', va='bottom')
        
    x = df_deal['trade_date'].values
    y1 = df_deal['TR'].values
    y2 = df_deal['close'].values
    y3 = df_deal['ATR'].values
    y4 = df_deal['pct_chg'].values
    
    plt.plot(x, y1, label='TR')
    plt.plot(x, y2, label='close')
    plt.plot(x, y3, label='ATR')
    plt.bar(x, y4, label='pct_chg')
    plt.xticks(rotation=70)
    plt.legend()
    return df_deal

df1 = pro.daily(ts_code='002547.SZ', start_date='20190801', end_date=today)[::-1]
df1_deal = showRes(df1, '002547.SZ', days=14)
df1_deal
```
